# Geocoding data
Data only had addresses but I wanted to show a density map of the bike stations.  
I initially tried to geocode each address one by one but with almost 30 million rows it would take forever.  
I decided to best represent the data and get the info I needed getting unique addresses, which accounted for under 600 would give me faster geocoded data and then just assigned the value counts to those addresses.  
I make a seperate csv file use to show density mapping using Tableau.  

```python
#Preprocessing address data
# Remove all characters except letters belonging to english alphabet, spaces and tabs

df['Start station'] = df['Start station'].str.replace('[^A-Za-z\s0-9]+', '')
df.head(10)
```

```python
#To get all unique stations
stationlist = list(df['Start station'].unique())
```

```python
#Show's value counts for each unique station
df['Start station'].value_counts
```

```python
df = pd.read_csv('201901-capitalbikeshare-tripdata.csv')
```

```python
#Added a count colume with everything equaling 1 since every row was one single rental.  
#Helps me keep track and get total counts.
df['count']=1
```

```python
#To get more accurate Geocoded results I added District of Columbia to all station addresses
df['Address_1'] = (df['Start station'] + ', District of Columbia')
df.head(10)
```

```python
#Gives me the value counts of each unique address
value_counts = df['Address_1'].value_counts()
```

```python
#Putting those value counts into a dataframe
df_val_counts = pd.DataFrame(value_counts)
```

```python
#Resetting index
df = df_val_counts.reset_index()
```

```python
#Now I have unique addresses in one column and the number of counts(rentals) in another column
#Doing this will help me when I geocode the results instead of processing millions of addresses I will be processing under 600 addresses
df.columns = ['Address_1', 'counts'] 
df.head()
```

```python
#Putting all addresses into a list
stationlist = list(df['Address_1'].unique())
```

```python
#Checking list
stationlist
```

```python
# Importing the necessary modules for this tutorial
# Pandas Library for fast, flexible, and expressive data structures designed

import pandas as pd
from geopy.geocoders import Nominatim, ArcGIS, GoogleV3 # Geocoder APIs
```

```python
g = Nominatim() # You can tryout ArcGIS or GoogleV3 APIs to compare the results
```

```python
#Testing if it works
# Geocoding - Address to lat/long

n = g.geocode('Empire State Building New York', timeout=10) # Address to geocode
print(n.latitude, n.longitude)
```

```python
add_list = [] # an empty list to hold the geocoded results

for add in stationlist:
    print ('Processing .... ', add)
    
    try:
        n = g.geocode(add, timeout=10)
        
        data = (add, n.latitude, n.longitude)
        add_list.append(data)
        
    except Exception:
        data = (add, "None", "None")
        add_list.append(data)
```

```python
# make a new dataframe to hold geocoded reult
#Import info I need is Lat and Long.  I will use these to get location in Tableau

add_list_df = pd.DataFrame(add_list, columns=['Start station', 'Latitude', 'Longitude'])
add_list_df.head(20)
```

```python
#Combining my unique addresses with counts and lat and long
newdf = pd.concat([add_list_df, df], axis = 1)
```

```python
#Testing if it works
newdf.head(20)
```

```python
#Geocoded results saved to a new csv, which I will import to Tableau
newdf.to_csv(r'geocoded.csv')
```
